```python
import datetime as dt
import pandas as pd
import time

def date_range(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + dt.timedelta(n)

start_date=dt.date(2019,9,20)
end_date=dt.date(2020,7,16)
for i in date_range(start_date, end_date):
    gf = pd.read_csv('E:\\DB103RichardC\\python\\190826PyETL\\dataHUB\\googleFlight\\depart\\2019-09-19\\google_flight_dep'+i.strftime("%Y-%m-%d")+'.csv')

    # abandon rows w/ missing values
    gf = gf.dropna()

    # filtering flights w/ transfer(s)
    for j in gf.index:
        if gf.loc[j,'dep_stops'] != 0:
            gf = gf.drop(index=j)
        else:
            pass
    gf = gf.drop(columns='dep_stops')

    # transfer time columns data type into correct ones
    gf['dep_etd']=pd.to_datetime(gf['dep_etd'])
    gf['dep_duration'] = [x + ':00' for x in gf['dep_duration']]
    gf['dep_duration']=pd.to_timedelta(gf['dep_duration'])
    gf = gf.drop(columns=['dep_eta'])
    gf.insert(3,'dep_eta',gf['dep_etd'] + gf['dep_duration'])

    # call airline ranking table
    rk = pd.read_csv('E:\\DB103RichardC\\python\\190826PyETL\\dataHUB\\googleFlight\\ar_clean_dep.csv', usecols=['dep_airline','score'])
    # merge the scores into major table
    gk = pd.merge(gf,rk,how='left', on='dep_airline')
    #save files
    gk.to_csv('E:\\DB103RichardC\\python\\190826PyETL\\dataHUB\\googleFlight\\depart\\2019-09-19cleaned\\dep'+i.strftime("%Y-%m-%d")+'.csv', index=False)
    print(i.strftime("%Y-%m-%d")+'file done. Mission runtime=', time.process_time())
```
