```python
import datetime as dt
import pandas as pd
import time

def date_range(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + dt.timedelta(n)

start_date=dt.date(2019,9,20)
end_date=dt.date(2020,7,16)
for i in date_range(start_date, end_date):
    gf = pd.read_csv('E:\\DB103RichardC\\python\\190826PyETL\\dataHUB\\googleFlight\\return\\2019-09-19\\google_flight_ret'+i.strftime("%Y-%m-%d")+'.csv')

    # abandon rows w/ missing values
    gf = gf.dropna()

    # filtering flights w/ transfer(s)
    for j in gf.index:
        if gf.loc[j,'ret_stops'] != 0:
            gf = gf.drop(index=j)
        else:
            pass
    gf = gf.drop(columns='ret_stops')

    # transfer time columns data type into correct ones
    gf['ret_etd']=pd.to_datetime(gf['ret_etd'])
    gf['ret_duration'] = [x + ':00' for x in gf['ret_duration']]
    gf['ret_dur']=pd.to_timedelta(gf['ret_duration'])
    gf = gf.drop(columns=['ret_eta'])
    gf.insert(3,'ret_eta',gf['ret_etd'] + gf['ret_dur'])
    gf = gf.drop(columns=['ret_dur'])
    # call airline ranking table
    rk = pd.read_csv('E:\\DB103RichardC\\python\\190826PyETL\\dataHUB\\googleFlight\\ar_clean_ret.csv', usecols=['ret_airline','score'])
    # merge the scores into major table
    gk = pd.merge(gf,rk,how='left', on='ret_airline')
    #save files
    gk.to_csv('E:\\DB103RichardC\\python\\190826PyETL\\dataHUB\\googleFlight\\return\\2019-09-19cleaned\\ret'+i.strftime("%Y-%m-%d")+'.csv',index=False)
    print(i.strftime("%Y-%m-%d")+'ret file done. Mission runtime=', time.process_time())
```
